```python
#import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import requests
import json
```

```python
#get URL
url='https://www.calories.info/food/fruits'
```

```python
#request URL
page=requests.get(url) 

#Check Status Code
page.status_code

#Parse Page Content
soup = BeautifulSoup(page.content, 'html.parser')

#Find Table 
fruits=soup.find('tbody')
#fruits=soup.find('table', class_='table dataTable no-footer calorie-hide-kj')

fruits
```

```python
# Contain scraped data in lists for dataframe
name=[]
per_serving=[]
calories=[]

# append for each list from rows in table 
for fruit in fruits:
    name.append(fruit.find('a').get_text())
    per_serving.append(fruit.find('td',class_="serving portion").get_text().replace('\xa0',''))
    calories.append(fruit.find('td',class_="kcal").get_text().replace("cal",""))
```

```python
#Include Lists in new Dataframe
df_1=pd.DataFrame({'Name':name,
                   'Per Serving (Grams)': per_serving,
                   'Calories (per 100g)': calories
                  })
```

```python
#Convert Calories to Int
df_1['Calories (per 100g)'] = df_1['Calories (per 100g)'].astype(int)
```

```python
#Parse json values into lists using for loop
carbohydrate=[]
sugar=[]
fiber=[]
protein=[]
fat=[]
Name=[]
cal=[]
for macro in data:
    carbohydrate.append(macro["carbohydrates_total_g"])
    sugar.append(macro["sugar_g"])
    fiber.append(macro["fiber_g"])
    protein.append(macro["protein_g"])
    fat.append(macro["fat_total_g"])
    Name.append(macro["name"])
    cal.append(macro["calories"])
```
